```python
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
class_names = ['Dog','Cat','Mouse']

def prepare(filepath):
    IMG_SIZE=50
    img_array=cv2.imread(filepath,cv2.IMREAD_GRAYSCALE)
    
    new_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
    return new_array.reshape(-1,IMG_SIZE,IMG_SIZE,1)

model=tf.keras.models.load_model("Dogs-vs-Cats_64x3_CNN.model")

newImage=input("Enter the images path: \n")
img_array=prepare(newImage)
img_array
prediction=model.predict([img_array])
predicted_class = class_names[np.argmax(prediction)]

print(f"The predicted class: {predicted_class}")

real_image=cv2.imread(newImage)
plt.imshow(real_image,cmap="gray")
plt.show()
```
