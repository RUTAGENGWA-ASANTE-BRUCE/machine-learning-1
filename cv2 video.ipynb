```python
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
class_names = ['paper','scissor','rock']


def prepare(frame):
    IMG_SIZE=70
    new_array=cv2.resize(frame,(IMG_SIZE,IMG_SIZE))
    return new_array.reshape(-1,IMG_SIZE,IMG_SIZE,1)

model=tf.keras.models.load_model("Paper_Scissor_Rock_CNN.model")

cap = cv2.VideoCapture(0)
cap.set(3,640) # adjust width
cap.set(4,480) # adjust height


while True:
    success, img = cap.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cv2.imshow("Webcam", gray) # This will open an independent window
    new_array=prepare(gray);
#     print(new_array.shape)
    prediction=model.predict([new_array])
    predicted_class = class_names[np.argmax(prediction)]
#     if(predicted_class=="paper"):
    print(f"The predicted class: {predicted_class}",end=" ")
        

    if cv2.waitKey(1) & 0xFF==ord('q'): # quit when 'q' is pressed
        cap.release()
        break
        
cv2.destroyAllWindows() 
cv2.waitKey(1) # no
```
